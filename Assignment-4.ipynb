
# Python List and Tuple Questions

### 1. What exactly is []?
`[]` represents an empty list in Python, which is a list with no elements.

### 2. In a list of values stored in a variable called `spam`, how would you assign the value `'hello'` as the third value? 
(Assume `[2, 4, 6, 8, 10]` are in `spam`.)

```python
spam = [2, 4, 6, 8, 10]
spam[2] = 'hello'
```

### 3. What is the value of `spam[int(int('3' * 2) / 11)]`? 
Let's pretend `spam` includes the list `['a', 'b', 'c', 'd']`.

The expression `int('3' * 2)` repeats the string `'3'` twice to get `'33'`, which is then converted to the integer `33`. Dividing `33` by `11` gives `3`. Thus, `spam[int(int('3' * 2) / 11)]` is `spam[3]`, which is `'d'`.

### 4. What is the value of `spam[-1]`? 
The value is `'d'` because `-1` refers to the last element of the list.

### 5. What is the value of `spam[:2]`? 
The value is `['a', 'b']`, as it slices the list to include elements from the start up to, but not including, index `2`.

### 6. What is the value of `bacon.index('cat')`? 
Let's pretend `bacon` has the list `[3.14, 'cat', 11, 'cat', True]`.

The value is `1`, as `index()` returns the index of the first occurrence of the value `'cat'`.

### 7. How does `bacon.append(99)` change the look of the list value in `bacon`? 

```python
bacon = [3.14, 'cat', 11, 'cat', True]
bacon.append(99)
# bacon is now [3.14, 'cat', 11, 'cat', True, 99]
```

### 8. How does `bacon.remove('cat')` change the look of the list in `bacon`? 

```python
bacon = [3.14, 'cat', 11, 'cat', True]
bacon.remove('cat')
# bacon is now [3.14, 11, 'cat', True]
```

### 9. What are the list concatenation and list replication operators? 
The list concatenation operator is `+`, and the list replication operator is `*`.

### 10. What is the difference between the list methods `append()` and `insert()`? 

- `append()` adds an item to the end of the list.
- `insert()` adds an item at a specified index.

### 11. What are the two methods for removing items from a list? 

- `remove()` removes the first occurrence of a value.
- `pop()` removes an item at a specified index or the last item if no index is specified.

### 12. Describe how list values and string values are identical.

Both lists and strings are sequences that can be indexed, sliced, and iterated over. Both can be concatenated and replicated, and they have similar methods for finding, replacing, and counting elements.

### 13. What's the difference between tuples and lists? 

- Lists are mutable (can be changed).
- Tuples are immutable (cannot be changed).

### 14. How do you type a tuple value that only contains the integer `42`?

```python
single_tuple = (42,)
```

### 15. How do you get a list value's tuple form? How do you get a tuple value's list form?

- Use `tuple()` to convert a list to a tuple.
- Use `list()` to convert a tuple to a list.

### 16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain? 

They contain references to the list values.

### 17. How do you distinguish between `copy.copy()` and `copy.deepcopy()`?

- `copy.copy()` creates a shallow copy of a list, meaning it only copies the references to the objects in the list.
- `copy.deepcopy()` creates a deep copy, meaning it copies the objects as well as any objects referenced by the objects in the list.
