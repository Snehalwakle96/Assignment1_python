```python
# 1. To what does a relative path refer?

"""
A relative path refers to a location that is relative to a current directory. Instead of specifying the complete path 
from the root directory, a relative path starts from the current working directory and leads to the target file or folder.
"""

# 2. What does an absolute path start with on your operating system?

"""
An absolute path starts with the root directory of your operating system. 
On Windows, it typically starts with a drive letter (e.g., C:\), and on Unix-based systems like Linux or macOS, 
it starts with a forward slash (/).
"""

# 3. What do the functions os.getcwd() and os.chdir() do?

"""
- os.getcwd(): This function returns the current working directory, which is the folder in which the Python script is running.
- os.chdir(): This function changes the current working directory to a specified path.
"""

import os

# Example:
current_dir = os.getcwd()  # Gets the current working directory
print("Current Directory:", current_dir)

# Change the working directory
new_dir = "/path/to/directory"
# os.chdir(new_dir)
# print("New Directory:", os.getcwd())

# 4. What are the . and .. folders?

"""
- . refers to the current directory.
- .. refers to the parent directory (one level up from the current directory).
"""

# 5. In C:\bacon\eggs\spam.txt, which part is the dir name, and which part is the base name?

"""
- Dir name: C:\bacon\eggs
- Base name: spam.txt
"""

path = "C:\\bacon\\eggs\\spam.txt"
dirname = os.path.dirname(path)
basename = os.path.basename(path)

print("Dir name:", dirname)
print("Base name:", basename)

# 6. What are the three “mode” arguments that can be passed to the open() function?

"""
The three common mode arguments that can be passed to the open() function are:
- 'r' for reading (default mode)
- 'w' for writing (truncates the file if it exists)
- 'a' for appending (writes data at the end of the file if it exists)
"""

# 7. What happens if an existing file is opened in write mode?

"""
If an existing file is opened in write mode ('w'), the file's contents are truncated (deleted), and a new empty file 
is created with the same name. Any existing data in the file is lost.
"""

# 8. How do you tell the difference between read() and readlines()?

"""
- read(): Reads the entire content of a file as a single string.
- readlines(): Reads the entire content of a file line by line and returns a list of strings, where each string is a line 
from the file.
"""

# Example:
file_path = "example.txt"

# with open(file_path, 'r') as file:
#     content = file.read()  # Reads entire content as a single string
#     print("Content with read():", content)

# with open(file_path, 'r') as file:
#     lines = file.readlines()  # Reads content line by line into a list
#     print("Content with readlines():", lines)

# 9. What data structure does a shelf value resemble?

"""
A shelf value resembles a dictionary. It allows you to store key-value pairs, where the keys are strings, 
and the values can be any Python object. It provides persistent storage for Python objects using the `shelve` module.
"""
```
